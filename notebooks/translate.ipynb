```python
from sarvamai import SarvamAI

import os
from dotenv import load_dotenv
```

```python
# Load Search API Key
load_dotenv()  # reads variables from a .env file and sets them in os.environ
```

```python
client = SarvamAI(
    api_subscription_key=os.environ["SARVAM_API_KEY"],
)
```

```python
def translate(input_text):

    response = client.text.translate(
        input= input_text,           # limit 1000 chars for mayura:v1 model
        source_language_code="auto",
        target_language_code="en-IN",
        speaker_gender="Female",
        mode="modern-colloquial",   # best for hinglish input, chatbot use cases
        model="mayura:v1",         # "sarvam-translate:v1" does not support modern colloquial
        numerals_format="international",
    )

    translated_text = response.translated_text

    return translated_text
```

```python
input_query = "मुझे पायथॉन भाषा समझाओ"

translate(input_query)
```

```python
input_query = "मेरे घूमने के लिए, भारत में कुछ मशहूर जगहें कौन-सी हैं?"

translate(input_query)
```
